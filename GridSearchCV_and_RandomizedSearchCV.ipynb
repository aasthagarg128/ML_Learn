#**GridSearchCV**

GridSearchCV is used for determining the best parameters for our model

````python
# loading the SVC
model=SVC()
````

````python
# hyperparameter

parameters ={
            'kernel':['linear','poly','rbf','sigmoid'],
            'C':[1,5,10,20]
}
````

Great question! Let's talk about what **`C`** means in SVM and Logistic Regression — in **simple, real-world terms** 🌟

---

## 🔧 What is `C`?

In both **Support Vector Machine (SVC)** and **Logistic Regression**, `C` is a **hyperparameter** that controls the **trade-off between accuracy on the training set and simplicity of the decision boundary**.

---

### ⚖️ Layman Explanation:

Think of `C` like a **strictness meter** for the model:

- **High `C` (e.g., 10, 100)**  
  → The model **focuses more** on getting **every training example correct**, even if that means a **complex boundary** (risk of overfitting).

- **Low `C` (e.g., 0.01, 1)**  
  → The model allows **some mistakes** on the training data in favor of a **simpler boundary** (better generalization).

---

### 📈 In SVM:
- A **smaller `C`** allows **slack** in the margin — letting some points be misclassified to keep the margin wide.
- A **larger `C`** forces the classifier to **fit the training data tightly** — possibly leading to overfitting.

### 📉 In Logistic Regression:
- **`C` is the inverse of regularization strength**.
  - So **lower `C` = stronger regularization** (simpler model).
  - **Higher `C` = weaker regularization** (more complex model).

---

### ✅ Example Grid Search (like your code):
```python
parameters = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [1, 5, 10, 20]
}
```

You're testing how the model performs with:
- different **types of boundaries** (kernel)
- different **strictness levels** (C)

---

## 🧠 Summary

| C Value | Meaning | Risk | Analogy |
|---------|---------|------|---------|
| Low     | Loose fit, allows errors | Underfitting | Chill teacher: "A few mistakes are okay!" |
| High    | Tight fit, fewer errors | Overfitting | Strict teacher: "No mistakes allowed!" |

---

````python
# grid search
classifier = GridSearchCV(model,parameters,cv=5)
````

````python
# fitting the data to our model
classifier.fit(X,Y)
````

````python
classifier.cv_results_
````

````python
# best parameters

best_parameters = classifier.best_params_
print(best_parameters)
````

````python
# highest accuracy

highest_accuracy = classifier.best_score_
print(highest_accuracy)
````

````python
# loading the results to pandas dataframe
result=pd.DataFrame(classifier.cv_results_)
````

````python
result.head()
````

````python
grid_search_result= result[['param_C','param_kernel','mean_test_score']]
````

````python
grid_search_result
````

Highest Accuracy= 95.2569%

Best Parameters = {'C':10
                  'Kernel': 'linear' }

#**RandomizedSearchCV**





````python
# loading the SVM model
model=SVC()
````

````python
# Hyperparameter

parameter={
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
````

````python
# Randomize search
classifier=RandomizedSearchCV(model,parameter,cv=5)
````

````python
# fitting the data to our model
classifier.fit(X,Y)
````

````python
classifier.cv_results_
````

````python
# best parameters

best_parameters = classifier.best_params_
print(best_parameters)
````

````python
highest_accuracy=classifier.best_score_
print(highest_accuracy)
````

````python
# loading the results to pandas dataframe
result=pd.DataFrame(classifier.cv_results_)
````

````python
result.head()
````

````python
randomized_search_result=result[['param_C','param_kernel','mean_test_score']]
````

````python
randomized_search_result
````

Highest Accuracy= 95.2569%

Best Parameters = {'C':10
                  'Kernel': 'linear' }
