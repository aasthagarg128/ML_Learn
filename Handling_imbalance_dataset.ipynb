Imbalanced Dataset:

A dataset  with an unequal class distribution

```python
# importing dependencies
import pandas as pd
import numpy as np
```

```python
# loading the dataset to the pandas  dataframe
cred_data = pd.read_csv('/content/credit_data.csv')
```

```python
cred_data.head()
```

```python
cred_data.tail()
```

```python
cred_data.shape
```

```python
# distribution of the 2 classes
cred_data['Class'].value_counts()
```

This is  HIGHLY IMBALANCED dataset

0 ---> Legit transactions


1---> Fraudulent transaction

```python
# separating thelegit and fraudulent transaction
legit = cred_data[cred_data.Class==0]
fraud = cred_data[cred_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**UNDER-SAMPLING**

Building  a sample dataset containing similar distributionof legit and fraudulent transactions

```python
legit_sample = legit.sample(n=442)
```

```python
print(legit_sample.shape)
```

Concatenate the two data frames

```python
new_dataset=pd.concat([legit_sample,fraud],axis = 0)
```

**Understanding axis in Pandas:**


axis=0 → Operates on rows (drop rows).


axis=1 → Operates on columns (drop columns)
