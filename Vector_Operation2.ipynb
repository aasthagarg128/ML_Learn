```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Dot Product of Two Vector




```python
a = np.asarray([2,3])
b = np.asarray([4,4])

a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
c = np.asarray([40,20,35])
d = np.asarray([53, 24, 68])

c_dot_d=np.dot(c,d)
print(c_dot_d)
```

Cross Product of Two Vector


```python
a = np.asarray([2,3])
b = np.asarray([4,4])

a_cross_b = np.cross(a,b)
print(a_cross_b)
```

```python
c = np.asarray([5,10,20])
d = np.asarray([18,32,50])

c_cross_d = np.cross(c,d)
print(c_cross_d)
```

Projection of "a" vector on "v" vector

```python
a = np.asarray([2,5])
v = np.asarray([8,-6])

# magnitude of "v" vector
magnitude_v = np.sqrt(sum(v**2))
a_dot_v = np.dot(a,v)

proj_a= a_dot_v/magnitude_v**2 * v

print(proj_a)
```
