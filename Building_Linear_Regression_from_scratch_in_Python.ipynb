```python
# importing libraries
import numpy as np
```

**Linear Regression**

```python
# class in python is used to create template for object

class Linear_Regression():
  # initiation function-used to initiate some parameters for learning model
  def __init__(self,learning_rate,no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations= no_of_iterations

  # It will be used to train the linear regression model with the  dataset we have
  def fit(self,X,Y): # x=years of exp , y=salary

    # m=no of training  examples(no.of data points) & n=number of features-can be multiple features
    self.m, self.n=X.shape  # number of rows and column

    # initiating the weight and bias
    self.w=np.zeros(self.n)  # creating an array with 0 as input where n = no. of features
    self.b=0
    self.X=X
    self.Y=Y

    # implementing Gradient Descent
    for i in range(self.no_of_iterations):
      self.update_weights()

  def update_weights(self, ):
    Y_prediction = self.predict(self.X)

    # calculate gradient

    dw=-(2*(self.X.T).dot(self.Y-Y_prediction))/self.m

    db=-2*np.sum(self.Y-Y_prediction)/self.m

    #updating the weights

    self.w=self.w-self.learning_rate*dw
    self.b=self.b-self.learning_rate*db


  def predict(self,X ):

    return X.dot(self.w)+self.b  # Y=wX+b
```
