Matplotlib:

*   Useful for making Plots, Graphs, Diagrams.



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plot
import numpy as np
```

```python
a = np.linspace(0,10,100)    # 0 to 10 are divided into 100 values
b = np.sin(a)
c = np.cos(a)
print(a)
print(b)
print(c)
```

Plotting the Data

```python
# sine wave
plt.plot(a,b)
plt.show()
```

```python
# cosine wave
plt.plot(a,c)
plt.show()
```

```python
# adding title, x-axis & y-axis labels
plt.plot(a,b)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
a = np.linspace(-10,10,20)
b = a**2
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,b,'r+')  # r represents red color and + symbol with with the graph gets plotted
plt.show()
```

```python
plt.plot(a,b,'g.')  # g represents green color and . symbol with with the graph gets plotted
plt.show()
```

```python
plt.plot(a,b,'yx')  # y represents yellow color and x symbol with with the graph gets plotted
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x),'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])  # 0,0 represents coordinates of origin  1,1 represents height and width of the rectangle
languages = ['Java', 'Python', 'c++', 'Ruby', 'golang']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('People')
plt.title('Popularity of Programming Languages')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
languages = ['Java', 'Python', 'c++', 'Ruby', 'golang']
people = [100,50,150,40,70]
ax1.pie(people , labels=languages , autopct='%1.1f%%')   # autopct='%1.1f%%' tells how many values are needed after decimal
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y,color='g')
ax2.scatter(x,z,color='r')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x= np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z, c=z ,cmap='Blues',)
plt.show()
```
